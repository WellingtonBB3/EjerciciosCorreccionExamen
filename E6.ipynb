```python
import numpy as np
import matplotlib.pyplot as plt

def Spline(x, x0, pars):
    """Función que evalúa un spline cúbico en x."""
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3

# Puntos dados
xs = [-1, 0, 1]
ys = [1, 5, 3]

# Pendiente deseada en el punto (x1, y1)
x1, y1 = 0, 5
m = 2  # Pendiente deseada

# Coeficientes iniciales de los splines
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]

# Ajustar la pendiente en (x1, y1)
# La pendiente en el spline debe ser igual a m en x1
# S'(x1) = b + 2 * c * (x1 - x0) + 3 * d * (x1 - x0)^2
s[0]["b"] = m - 2 * s[0]["c"] * (x1 - xs[0]) - 3 * s[0]["d"] * (x1 - xs[0]) ** 2

# Graficar los splines ajustados
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 50)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, label=f"Spline {i+1}")

# Dibujar los puntos originales
plt.scatter(xs, ys, color="black", zorder=5, label="Puntos originales")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Splines cúbicos con pendiente ajustada")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función del spline cúbico
def Spline(x, x0, pars):
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0)**2 + d * (x - x0)**3

# Puntos originales
xs = [-1, 0, 1]
ys = [1, 5, 3]

# Coeficientes de los splines cúbicos para cada intervalo
splines = [
    {"a": 1, "b": 7, "c": 0, "d": -3},  # Spline 1 (intervalo [-1, 0])
    {"a": 5, "b": -2, "c": 0, "d": 0},  # Spline 2 (intervalo [0, 1])
]

# Graficar cada spline en su intervalo correspondiente
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 100)  # Generar puntos en el intervalo
    _y = Spline(_x, x_i, splines[i])  # Evaluar el spline en esos puntos
    plt.plot(_x, _y, label=f"Spline {i + 1}")  # Graficar el spline

# Graficar los puntos originales
plt.scatter(xs, ys, color="black", zorder=5, label="Puntos originales")

# Etiquetas y título
plt.xlabel("x")
plt.ylabel("y")
plt.title("Splines cúbicos con pendiente ajustada")
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()
```
